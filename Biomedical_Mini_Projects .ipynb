```python
import wfdb
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks

# load ECG record
record = wfdb.rdrecord('100', pn_dir='mitdb')
signal = record.p_signal[:,0]

# detect heartbeats
peaks, _ = find_peaks(signal, distance=200)

# plot
plt.figure(figsize=(10,4))
plt.plot(signal[:2000])
plt.plot(peaks, signal[peaks], "ro")
plt.title("Detected Heartbeats in ECG Signal")
plt.xlabel("Time")
plt.ylabel("Voltage")
plt.show()

print("Number of detected heartbeats:", len(peaks))
```
